### Resolucao do Problema 12

```python
from problemas_selecionados import problema12 as p12
fobj = p12.func
lb,ub = p12.constraint()
# Parametros para todos os metodos
N_max = 500
n = 30

all_best1 = []
all_fmin1 = []
all_conv_curve1 = []

# Executando o loop 30 vezes
start_time31 = time.time()
for _ in range(30):
    # Parametros para o DE
    F = 0.7
    Cr = 0.9
    best1,fmin1,melhor_curve1 = myDE(fobj,n,F,Cr,lb,ub,N_max)
    all_best1.append(best1)
    all_fmin1.append(fmin1)
    all_conv_curve1.append(melhor_curve1)
end_time31 = time.time()

all_fmin1_array = np.array(all_fmin1) # Transforme all_best1 em um array numpy para facilitar os cálculos
pior1 = np.max(all_fmin1_array) # Pior resultado (maior valor)
fmelhor1 = np.min(all_fmin1_array) # Melhor resultado (menor valor)
media1 = np.mean(all_fmin1_array) # Média
desvpad1 = np.std(all_fmin1_array) # Desvio padrão
indice1 = np.argmin(all_fmin1_array) # Índice da melhor solução (menor valor)
melhor_solucao1 = all_best1[indice1] # Melhor solução
conv_curve1 = all_conv_curve1[indice1] # Curva correspondente à melhor solução

print("Solucao por Evolucao Diferencial") 
print(f"Ponto Otimo: {melhor_solucao1}")
print(f"Solucao Otima: {fmelhor1}")
print(f"Media: {media1}")
print(f"Pior: {pior1}")
print(f"DesvioPadrao: {desvpad1}")
print(f"Tempo de execução: {(end_time31-start_time31)/30:.4f} segundos")
######################################################################
######################################################################
######################################################################
######################################################################
# Parametros para o PSO
c1 = 2 # constante cognitiva
c2 = 2 # constante social
all_best2 = []
all_fmin2 = []
all_conv_curve2 = []

start_time32 = time.time()
for _ in range(30):
    best2,fmin2,melhor_curve2 = myPSO(fobj,n,c1,c2,lb,ub,N_max)
    all_best2.append(best2)
    all_fmin2.append(fmin2)
    all_conv_curve2.append(melhor_curve2)
end_time32 = time.time()

all_fmin2_array = np.array(all_fmin2) # Transforme all_best1 em um array numpy para facilitar os cálculos
pior2 = np.max(all_fmin2_array) # Pior resultado (maior valor)
fmelhor2 = np.min(all_fmin2_array) # Melhor resultado (menor valor)
media2 = np.mean(all_fmin2_array) # Média
desvpad2 = np.std(all_fmin2_array) # Desvio padrão
indice2 = np.argmin(all_fmin2_array) # Índice da melhor solução (menor valor)
melhor_solucao2 = all_best2[indice2] # Melhor solução
conv_curve2 = all_conv_curve2[indice2] # Curva correspondente à melhor solução

print("\nSolucao por Enxame de Particulas")
print(f"Ponto Otimo: {melhor_solucao2}")
print(f"Solucao Otima: {fmelhor2}")
print(f"Media: {media2}")
print(f"Pior: {pior2}")
print(f"DesvioPadrao: {desvpad2}")
print(f"Tempo de execução: {(end_time32-start_time32)/30:.4f} segundos")
######################################################################
######################################################################

######################################################################
######################################################################
# Parametros para o AOA
#C3=2;C4=.5;
C3 = 1; C4 =2;

all_best5 = []
all_fmin5 = []
all_conv_curve5 = []
start_time33 = time.time()
for _ in range(30):
    best5,fmin5,melhor_curve5 = myAOA(fobj,n,N_max,lb,ub,C3,C4)
    all_best5.append(best5)
    all_fmin5.append(fmin5)
    all_conv_curve5.append(melhor_curve5)
end_time33 = time.time()

all_fmin5_array = np.array(all_fmin5) # Transforme all_best1 em um array numpy para facilitar os cálculos
pior5 = np.max(all_fmin5_array) # Pior resultado (maior valor)
fmelhor5 = np.min(all_fmin5_array) # Melhor resultado (menor valor)
media5 = np.mean(all_fmin5_array) # Média
desvpad5 = np.std(all_fmin5_array) # Desvio padrão
indice5 = np.argmin(all_fmin5_array) # Índice da melhor solução (menor valor)
melhor_solucao5 = all_best5[indice5] # Melhor solução
conv_curve5 = all_conv_curve5[indice5] # Curva correspondente à melhor solução

print("\nSolucao por Algoritmo de Otimizacao de Arquimedes")
print(f"Ponto Otimo: {melhor_solucao5}")
print(f"Solucao Otima: {fmelhor5}")
print(f"Media: {media5}")
print(f"Pior: {pior5}")
print(f"DesvioPadrao: {desvpad5}")
print(f"Tempo de execução: {(end_time33-start_time33)/30:.4f} segundos")
######################################################################
######################################################################

######################################################################
######################################################################
# Chame a função SCA
all_best4 = []
all_fmin4 = []
all_conv_curve4 = []
start_time34 = time.time()
for _ in range(30):
    fmin4,best4,melhor_curve4 = mySCA(n, N_max,lb, ub, fobj)
    all_best4.append(best4)
    all_fmin4.append(fmin4)
    all_conv_curve4.append(melhor_curve4)
end_time34 = time.time()

all_fmin4_array = np.array(all_fmin4) # Transforme all_best1 em um array numpy para facilitar os cálculos
pior4 = np.max(all_fmin4_array) # Pior resultado (maior valor)
fmelhor4 = np.min(all_fmin4_array) # Melhor resultado (menor valor)
media4 = np.mean(all_fmin4_array) # Média
desvpad4 = np.std(all_fmin4_array) # Desvio padrão
indice4 = np.argmin(all_fmin4_array) # Índice da melhor solução (menor valor)
melhor_solucao4 = all_best4[indice4] # Melhor solução
conv_curve4 = all_conv_curve4[indice4] # Curva correspondente à melhor solução

print("\nSolucao por Algoritmo Seno Cosseno")
print(f"Ponto Otimo: {melhor_solucao4}")
print(f"Solucao Otima: {fmelhor4}")
print(f"Media: {media4}")
print(f"Pior: {pior4}")
print(f"DesvioPadrao: {desvpad4}")
print(f"Tempo de execução: {(end_time34-start_time34)/30:.4f} segundos")
######################################################################
######################################################################
```
